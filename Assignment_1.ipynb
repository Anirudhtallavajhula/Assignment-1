```python
import pandas as pd

# Create a dataframe with the given data
data = {
    "Height": [65.8, 71.5, 69.4, 68.2, 67.8, 68.7, 69.8, 70.1, 67.9, 66.8],
    "Weight": [112, 136, 153, 142, 144, 123, 141, 136, 112, 120],
    "Age": [30, 19, 45, 22, 29, 50, 51, 23, 17, 39],
    "Grip_strength": [30, 31, 29, 28, 24, 26, 22, 20, 19, 31],
    "Frailty": ["N", "N", "N", "Y", "Y", "N", "Y", "Y", "N", "N"]
}

# Convert to DataFrame and save it as a CSV file
df = pd.DataFrame(data)
df.to_csv("project/data/raw_data/female_frailty_data.csv", index=False)

df.head()  # Display the data
```

```python
# Load the raw data
df = pd.read_csv("project/data/raw_data/female_frailty_data.csv")

# Check for missing values
print(df.isnull().sum())

# Save cleaned data in the processed_data folder
df.to_csv("project/data/processed_data/cleaned_frailty_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt

# Descriptive statistics
df.describe()

# Analyze the relationship between grip strength and frailty
mean_grip_strength = df.groupby('Frailty')['Grip_strength'].mean()
print(mean_grip_strength)

# Visualize the data
plt.figure(figsize=(8,6))
plt.bar(mean_grip_strength.index, mean_grip_strength.values)
plt.title('Average Grip Strength by Frailty Status')
plt.xlabel('Frailty')
plt.ylabel('Average Grip Strength (kg)')
plt.savefig("project/analysis/figures/grip_strength_vs_frailty.png")
plt.show()
```

```python
# Save summary statistics to the results folder
summary = df.describe()
summary.to_csv("project/results/summary_statistics.csv")

# Save analysis report
with open('project/results/report.txt', 'w') as f:
    f.write("Grip Strength Analysis by Frailty Status\n\n")
    f.write(f"Mean Grip Strength (No Frailty): {mean_grip_strength['N']}\n")
    f.write(f"Mean Grip Strength (Frailty): {mean_grip_strength['Y']}\n")
```

```python
import matplotlib.pyplot as plt

# Histogram of math scores
plt.figure(figsize=(10, 6))
plt.hist(df['math score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

The distribution of arithmetic test results among kids is made clearer by this histogram. It enables us to find typical score ranges, identify any data skewness, and locate any outliers. Understanding the general arithmetic performance of pupils may be gained from this.

```python
import seaborn as sns

# Box plot of math scores by gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```

We are able to compare the distribution of arithmetic results between male and female students using the box plot. For each gender, the median, quartiles, and possible outliers are highlighted. This graphic facilitates the assessment of variability within each group and the identification of any gender-based performance disparities.


```python
# Bar plot of average math scores by parental education level
plt.figure(figsize=(10, 6))
avg_scores = df.groupby('parental level of education')['math score'].mean().reset_index()
sns.barplot(x='parental level of education', y='math score', data=avg_scores, palette='viridis')
plt.title('Average Math Scores by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Average Math Score')
plt.xticks(rotation=45)
plt.show()
```

Students' average arithmetic results are displayed in this bar plot according to the educational attainment of their parents. It sheds light on the possible relationship between parental education and student achievement. We can find any trends or patterns pertaining to parental education by comparing the average results across various education levels.


```python
# Violin plot of reading scores by test preparation course
plt.figure(figsize=(10, 6))
sns.violinplot(x='test preparation course', y='reading score', data=df, palette='muted')
plt.title('Reading Scores by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Reading Score')
plt.show()
```

The distribution of reading scores for students who finished the exam preparation course against those who did not can be shown in depth using the violin plot. It illustrates the distribution's shape and variability by fusing elements of a density plot with a box plot. We can better comprehend how test preparation affects reading results thanks to this depiction.


```python
# Pair plot of scores
plt.figure(figsize=(10, 6))
sns.pairplot(df[['math score', 'reading score', 'writing score']], diag_kind='kde', markers='+')
plt.suptitle('Pair Plot of Scores', y=1.02)
plt.show()
```

We may see how the scores in writing, reading, and math relate to one another with the pair plot. It has kernel density estimates (KDE) on the diagonal and scatter plots for every pair of scores. With the use of this visualization, we can see possible trends and relationships between the various score kinds, giving us a thorough understanding of how well students do throughout topic areas.
